# test train split

```python
# for learning purpose
import numpy as np
def split_train_test(data,test_ratio):
    np.random.seed(42)
    shuffled=np.random.permutation(len(data))
    print(shuffled)
    test_set_size=int(len(data)*test_ratio)
    test_indices=shuffled[:test_set_size]
    train_indices=shuffled[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

```python
# train_set,test_set=split_train_test(housing,.2)
```

# stratified shuffled split

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=.2,random_state=42)
for train_index,test_index in split.split(housing,housing['CHAS']):
    strat_train_set=housing.loc[train_index]
    strat_test_set=housing.loc[test_index]
```

```python
housing=strat_train_set.drop("MEDV",axis=1)
housing_labels=strat_train_set["MEDV"].copy()
#label ane data alg karva
```

# sklearn thi direct missing values fill karse

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
imputer.fit(housing)
```

```python
imputer.statistics_
```

```python
x=imputer.transform(housing)
```

```python
housing_tr=pd.DataFrame(x,columns=housing.columns)
```

```python
housing
```

# scikit learn design

creating pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline=Pipeline(
[('imputer',SimpleImputer(strategy='median')),
  ('standerd_scaler',StandardScaler()),
])
```

```python
housing_num_tr=my_pipeline.fit_transform(housing)
```

```python
housing_num_tr
```

# select and train model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

#model=LinearRegression()
# model=DecisionTreeRegressor()
model=RandomForestRegressor()
model.fit(housing_num_tr,housing_labels)
```

```python
somedata=housing.iloc[:10]
```

```python
somelabel=housing_labels.iloc[:10]
```

```python
prepared_data=my_pipeline.fit_transform(somedata)
```

```python
model.predict(prepared_data)
```

```python
somelabel
```

# evaluating the model

```python
from sklearn.metrics import mean_squared_error
housing_predict=model.predict(prepared_data)
lin_msc=mean_squared_error(housing_predict,somelabel)
lin_rmsc=np.sqrt(lin_msc)
lin_rmsc
lin_msc
```

# using better evaluation technique cross validattion

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,housing_num_tr,housing_labels,scoring='neg_mean_squared_error',cv=10)
```

```python
rmse_scores=np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(sc):
    print("scores:",sc)
    print("mean :",sc.mean())
    print("std :",sc.std())
```

```python
print_scores(rmse_scores)
```

```python
x_test=strat_test_set.drop('MEDV',axis=1)
y_test=strat_test_set['MEDV'].copy()
x_testp=my_pipeline.fit_transform(x_test)
final_predictions=model.predict(x_testp)
final_mse=mean_squared_error(y_test,final_predictions)
final_rmse=np.sqrt(final_mse)
```
