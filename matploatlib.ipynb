```python
import matplotlib.pyplot as plt
%matplotlib inline
#ide ma alg window open karse jupyter notbook ma kai farak nai pade
import numpy as np
x=np.arange(1,10,.1)
y=2*x+5
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,10,1)
y=2*x+5
y1=5*x+7
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,linewidth=2.0,linestyle=":",color="blue",alpha=.5,marker='o')
plt.plot(x,y1,linewidth=1.5,linestyle=":",color="orange",alpha=.8,marker='x')
plt.title("Line plot demo")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(['line1','line2'])
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,10,1)
y=2*x+5
y1=5*x+7

#create two or more plot subplot is used parameter ma first row column ane third kya plot rese te ave
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Graph1')

plt.subplot(1,2,2)
plt.plot(x,y1)
plt.title("Graph2")
```

```python
import matplotlib.pyplot as plt
import numpy as np
data={'apple':40,'banana':15,'kivy':70,'mango':20}
name=list(data.keys())
values=list(data.values())
fig=plt.figure(figsize=(10,5))
plt.bar(name,values)
plt.show()
```

```python
import matplotlib.pyplot as plt
data={'apple':40,'banana':15,'kivy':70,'mango':20}
name=list(data.keys())
values=list(data.values())
fig=plt.figure(figsize=(10,5))
plt.barh(name,values,color="orange",alpha=.6)
plt.title("Bar graph hoeizontal")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.show()
```

```python
import matplotlib.pyplot as plt
a=[10,20,30,40,50,60,70,80]
b=[4,7,8,3,4,7,1,5]
c=[1,8,4,6,3,5,7,8]
plt.scatter(a,b,color="green",s=200,edgecolor='black',marker='o',alpha=1)
plt.scatter(a,c,c='maroon',s=400,edgecolor='b',marker='4',alpha=.9)
plt.title("Scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(['b','c'])
plt.grid(True)
plt.savefig('sctterplot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

total=[20,4,1,30,20,10,20,70,30,10]
order=[10,3,1,15,17,2,30,44,2,1]
discount=[30,10,20,5,10,20,50,60,20,45]
data=list([total,order,discount])
plt.boxplot(data,showmeans=True)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

total=[20,4,1,30,20,10,20,70,30,10]
order=[10,3,1,15,17,2,30,44,2,1]
discount=[30,10,20,5,10,20,50,60,20,45]
data=list([total,order,discount])
plt.violinplot(data,showmeans=True,showmedians=True)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
label=['dog','cat','fox','lion']
sizes=[40,20,50,90]
plt.pie(sizes,labels=label,autopct='%1.1f%%',shadow=True,startangle=90,explode=(.1,.1,0,0))
plt.title("pie chart")
plt.show()
```
