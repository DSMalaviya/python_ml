# data visulization

```python
import matplotlib.pyplot as plt
y=cars['hp']
x=range(32)
plt.plot(x,y)
```

```python
%matplotlib inline
y1=cars['hp']
y2=cars['disp']
x=range(32)
plt.plot(x,y1,label="hp")
plt.plot(x,y2,label="disp")
plt.legend()
```

```python
y1=cars['hp']
y2=cars['disp']
x=range(32)
plt.plot(x,y1,label="hp")
plt.stackplot(x,y1,color="blue",alpha=.7)
plt.plot(x,y2,label="disp")
plt.stackplot(x,y2,color="green",alpha=.5)
plt.legend()
```

```python
y=cars['hp']
x=cars['model'].tolist()
fig=plt.figure(figsize=(50,20))
#to avoid issue of overlapping in barplot
plt.bar(x,y,color="blue",alpha=.2)
```

```python
y=cars['hp']
x=cars['model'].tolist()
fig=plt.figure(figsize=(50,20))
#to avoid issue of overlapping in barplot
plt.barh(x,y,color="blue",alpha=.8)
```
